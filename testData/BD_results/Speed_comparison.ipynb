# Plot the runtimes for Rustody and compare them to the BD system.

```python
import matplotlib.pyplot as plt

# Sample data
x = ['BD v1 AWS', 'BD v2 AWS', 'Rustody 48 cores', 'Rustody 30 cores', 'Rustody 20 cores', 'Rustody 10 cores', 'Rustody 5 cores']
y = [ 60*7 + 7, 60*1+23, 22, 23, 25, 31, 46]

colors = ['blue', 'blue', 'green', 'green', 'green', 'green', 'green']
# Create bar plot
plt.bar(x, y, color= colors)

# Add labels and title
plt.xlabel('')
plt.ylabel('Runtime [min]')
plt.title('Runtime comparison BD 7 bridges vs Rustody v2.1.0')


# Rotate x-axis labels
plt.xticks(rotation=90)

# Adjust layout to add space for x-axis labels
plt.tight_layout()

plt.savefig('bar_plot.svg')

# Show plot
plt.show()
```
